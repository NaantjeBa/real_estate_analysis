```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import HTML
import matplotlib
import numpy as np
```

```python
df_verhuurd = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\verhuurde_woningen_driemaanden.csv', sep=';')
df_verkocht = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\verkochte_woningen_driemaanden.csv', sep=';')
df_stadsdelen = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\dim_stadsdelen3.csv', sep=';')
```

```python
df_verhuurd['Huisnr.'] = df_verhuurd['Huisnr.'].fillna(0).astype(int).astype(str)
df_verhuurd['Toev.'] = df_verhuurd['Toev.'].fillna('')

df_verhuurd['id'] = df_verhuurd.Straat + "-" + \
df_verhuurd['Huisnr.'].astype(int).astype(str) + "-" + \
df_verhuurd['Toev.']

col_name_dict = {'Huidige prijs' : 'huidige_prijs_number', 
                 'Prijs per m2' : 'prijs_per_m2_number' ,
                'Transactie prijs / huur' : 'transactieprijs_huur_number', 
                 'Huidige huur' : 'huidige_huur_number',
                'Transactieprijs per m2' : 'transactieprijs_per_m2_number'}

"Convert euro column to type number"
for cur_col, new_col in col_name_dict.items():

    df_verhuurd[new_col] = df_verhuurd[cur_col].str.replace('€', '') \
    .str.replace(',', '')
    df_verhuurd[new_col] = pd.to_numeric(df_verhuurd[new_col], errors='coerce')

"Extract number from Postcode column"
df_verhuurd.loc[:, 'Postcode_cijfers'] = df_verhuurd.Postcode.str[:4].astype(int).copy()

cols = ['id', 
        'Postcode_cijfers', 
        #'Makelaar', 
        'huidige_prijs_number', 
        'Koop conditie', 
        'prijs_per_m2_number', 
        'transactieprijs_huur_number', 
        'transactieprijs_per_m2_number', 
        'huidige_huur_number', 
        'Huur conditie', 
        'Datum aanmelding',
       'Transactiedatum', 
        #'Aantal kamers', 
        'Soort OG', 
        #'Soort appartement', 
        'Status']

df_verhuurd_sub = df_verhuurd[cols]
dateformat = '%d-%m-%Y'
df_verhuurd_sub['Transactiedatum'] = pd.to_datetime(df_verhuurd_sub['Transactiedatum'], format=dateformat)
df_verhuurd_sub['Datum aanmelding'] = pd.to_datetime(df_verhuurd_sub['Datum aanmelding'], format=dateformat)

df_verhuurd_sub = df_verhuurd_sub.merge(df_stadsdelen, left_on='Postcode_cijfers', right_on='postcode', how='left')
```

```python
df_verkocht['Huisnr.'] = df_verkocht['Huisnr.'].fillna(0).astype(int).astype(str)
df_verkocht['Toev.'] = df_verkocht['Toev.'].fillna('')

df_verkocht['id'] = df_verkocht.Straat + "-" + \
df_verkocht['Huisnr.'].astype(int).astype(str) + "-" + \
df_verkocht['Toev.']

col_name_dict = {'Huidige prijs' : 'huidige_prijs_number', 
                 'Prijs per m2' : 'prijs_per_m2_number' ,
                'Transactie prijs / huur' : 'transactieprijs_huur_number', 
                 'Huidige huur' : 'huidige_huur_number',
                'Transactieprijs per m2' : 'transactieprijs_per_m2_number'}

"Convert euro column to type number"
for cur_col, new_col in col_name_dict.items():

    df_verkocht[new_col] = df_verkocht[cur_col].str.replace('€', '') \
    .str.replace(',', '')
    df_verkocht[new_col] = pd.to_numeric(df_verkocht[new_col], errors='coerce')

"Extract number from Postcode column"
df_verkocht.loc[:, 'Postcode_cijfers'] = df_verkocht.Postcode.str[:4].astype(int).copy()

cols = ['id', 
        'Postcode_cijfers', 
        #'Makelaar', 
        'huidige_prijs_number', 
        'Koop conditie', 
        'prijs_per_m2_number', 
        'transactieprijs_huur_number', 
        'transactieprijs_per_m2_number', 
        'huidige_huur_number', 
        'Huur conditie', 
        'Datum aanmelding',
       'Transactiedatum', 
        #'Aantal kamers', 
        'Soort OG', 
        #'Soort appartement', 
        'Status']

df_verkocht_sub = df_verkocht[cols]
dateformat = '%d-%m-%Y'
df_verkocht_sub['Transactiedatum'] = pd.to_datetime(df_verkocht_sub['Transactiedatum'], format=dateformat)
df_verkocht_sub['Datum aanmelding'] = pd.to_datetime(df_verkocht_sub['Datum aanmelding'], format=dateformat)

df_verkocht_sub = df_verkocht_sub.merge(df_stadsdelen, left_on='Postcode_cijfers', right_on='postcode', how='left')
```

```python
df_verkocht_sub = df_verkocht_sub[df_verkocht_sub['Koop conditie'] != 'Vrij op naam']
```

```python
df_verkocht_groupby = round(df_verkocht_sub.groupby(['stadsdeel', 'Soort OG'])['transactieprijs_per_m2_number'].agg(['mean']), 2)
df_verkocht_groupby.unstack()
```

```python
df_verkocht_mean_stadsdeel = round(df_verkocht_sub.groupby(['stadsdeel'])['transactieprijs_per_m2_number'].agg(['mean']), 2)
df_verkocht_mean_stadsdeel.columns = ['gem_verkoop_m2']
df_verkocht_mean_stadsdeel
```

```python
df_verhuurd_mean_stadsdeel = round(df_verhuurd_sub.groupby(['stadsdeel'])['transactieprijs_per_m2_number'].agg(['mean']) * 12 , 2)
df_verhuurd_mean_stadsdeel.columns = ['gem_verhuur_m2']
df_verhuurd_mean_stadsdeel
```

```python
df_list = [df_verkocht_mean_stadsdeel, df_verhuurd_mean_stadsdeel]
df_verhuur_verkoop = pd.concat(df_list, axis=1, sort=True)
df_verhuur_verkoop
```

```python
df_verhuur_verkoop['terug_verdien_jaren'] = round(df_verhuur_verkoop.gem_verkoop_m2 / df_verhuur_verkoop.gem_verhuur_m2, 1)
```

```python
df_verhuur_verkoop.sort_values('terug_verdien_jaren', inplace=True)
```
