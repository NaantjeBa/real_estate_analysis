```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import HTML
import matplotlib
import numpy as np
import data_prep
```

```python
df_verhuurd = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\verhuurde_woningen_driemaanden.csv', sep=';')
df_verkocht = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\verkochte_woningen_driemaanden.csv', sep=';')
df_stadsdelen = pd.read_csv('C:\\Users\\jniens\\GoogleDrive\\Python_Projects\\real_estate_analysis\\data_files\\csv\\dim_stadsdelen3.csv', sep=';')
```

```python
df_verhuurd_1 = data_prep.prep_df(df_verhuurd, df_stadsdelen)
df_verkocht_1 = data_prep.prep_df(df_verkocht, df_stadsdelen)
```

```python
df_verkocht_2 = \
df_verkocht_1.groupby(['stadsdeel', 'Soort OG'])['transactieprijs_perm2'] \
.agg(['mean']) \
.round(2) \
.unstack()
```

```python
"Remove multi level index"
level0 = df_verkocht_2.columns.get_level_values(0)
level1 = df_verkocht_2.columns.get_level_values(1)
df_verkocht_2.columns = level0 + '_' + level1
```

```python
df_verkocht_2['woonhuis_tov_appartement'] = ( ( df_verkocht_2.mean_Woonhuis / df_verkocht_2.mean_Appartement ) - 1 ) * 100
```

```python
"Show"
df_verkocht_2[['mean_Appartement', 'mean_Woonhuis']] \
.sort_values('mean_Woonhuis', ascending=False) \
.plot(kind='bar')

plt.show()
```

```python
"Make gem_verkoop_perm2"
df_verkocht_3 = df_verkocht_1.groupby(['stadsdeel'])['transactieprijs_perm2'] \
.agg(['mean']) \
.round(2)
df_verkocht_3.columns = ['gem_verkoop_perm2']

"Make gem_verhuur_perm2"
df_verhuurd_3 = df_verhuurd_1.groupby(['stadsdeel'])['transactieprijs_perm2'] \
.agg(['mean']) \
.mul(12) \
.round(2)
df_verhuurd_3.columns = ['gem_verhuur_perm2']
```

```python
"Concat the two dfs"
df_list = [df_verkocht_3, df_verhuurd_3]
df_verhuur_verkoop = pd.concat(df_list, axis=1, sort=True)

"Calculate terug_verdien_jaren"
df_verhuur_verkoop['terug_verdien_jaren'] = (df_verhuur_verkoop.gem_verkoop_perm2 / df_verhuur_verkoop.gem_verhuur_perm2).round(2)
df_verhuur_verkoop.sort_values('terug_verdien_jaren', inplace=True)
```

```python
cols = ['Status', 'Transactiedatum', 'transactieprijs_perm2']
df_verkocht_4 = df_verkocht_1[cols]
df_verhuurd_4 = df_verhuurd_1[cols]
```

```python
df_verkocht_verhuurd_4a = pd.concat([df_verkocht_4, df_verhuurd_4])
df_verkocht_verhuurd_4b = df_verkocht_verhuurd_4a.set_index('Transactiedatum').sort_index(ascending=True)
df_verkocht_verhuurd_4c = df_verkocht_verhuurd_4b.groupby(['Status', pd.Grouper(freq='W')]) \
.mean() \
.round(2) \
.unstack('Status') \
.sort_index()
df_verkocht_verhuurd_4c.columns = df_verkocht_verhuurd_4c.columns.droplevel()
df_verkocht_verhuurd_4c['Verhuurd'] = df_verkocht_verhuurd_4c.Verhuurd * 12
df_verkocht_verhuurd_4c['terugverdien_jaren'] = ( df_verkocht_verhuurd_4c.Verkocht / df_verkocht_verhuurd_4c.Verhuurd ).round(2)
df_verkocht_verhuurd_4d = df_verkocht_verhuurd_4c.loc['2019-04-28':]
```

```python
verhuurd_values = df_verkocht_verhuurd_4d.Verhuurd.values
verkocht_values = df_verkocht_verhuurd_4d.Verkocht.values
tvj_values = df_verkocht_verhuurd_4d.terugverdien_jaren.values
week_values = df_verkocht_verhuurd_4d.index.values
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

color1 = 'tab:red'
ax1.plot(week_values, verhuurd_values, color=color1)
ax1.set_ylabel('Verhuurprijs per m2', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)


color2 = 'tab:blue'
ax2 = ax1.twinx()
ax2.plot(week_values, verkocht_values, color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
ax2.set_ylabel('Verkoopprijs per m2', color=color2)

# color3 = 'tab:green'
# ax3 = ax1.twinx()
# ax3.plot(week_values, tvj_values, color=color3)
# ax3.tick_params(axis='y', labelcolor=color3)
# ax3.set_ylabel('Verkoopprijs per m2', color=color3)

fig.tight_layout()
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))

df_verkocht_5 = df_verkocht_1.copy()
df_verkocht_5.groupby(['stadsdeel'])[[ 'trans_gt_huid','trans_eq_huid','trans_lt_huid']] \
.mean() \
.sort_values('trans_gt_huid', ascending=False) \
.mul(100) \
.plot(kind='bar', ax=ax1, rot=45)
ax1.legend(['Transactieprijs hoger dan vraagprijs',
            'Transactieprijs gelijk aan vraagprijs',
            'Transactieprijs lager dan vraagprijs'])
ax1.set_xlabel('Stadsdeel', size=15)
ax1.set_ylabel('Percentage', size=15)
for label in ax1.get_xticklabels():
    label.set_horizontalalignment('right')
ax1.set_title('Vraagprijs versus Transactieprijs (Koopwoningen)\n', size=20)
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(14, 7))


df_verkocht_5b = df_verkocht_5.groupby(['stadsdeel'])[['prijs_perm2','transactieprijs_perm2']].mean().round(2)
df_verkocht_5b['trans_tov_huid'] = (( df_verkocht_5b.transactieprijs_perm2 - df_verkocht_5b.prijs_perm2 ) / df_verkocht_5b.prijs_perm2 * 100 ).round(2)
df_verkocht_5b.sort_values('trans_tov_huid', ascending=False, inplace=True)
df_verkocht_5b['trans_tov_huid'].plot(kind='bar', ax=ax1, rot=45)

ax1.set_xlabel('Stadsdeel', size=15)
ax1.set_ylabel('Percentage', size=15)
for label in ax1.get_xticklabels():
    label.set_horizontalalignment('right')
ax1.set_title('Verschil Vraagprijs versus Transactieprijs (Koopwoningen)\n', size=20)
plt.show()
```
